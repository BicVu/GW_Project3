```python
read_df = pd.read_csv('Clean_Data/ML_data_CLEAN.csv', encoding='utf-8')
pop_df = pd.read_csv('Clean_Data/pop_2000_2018.csv', encoding='utf-8')
```

```python
year = 2010
dg_df = read_df[read_df.Year.isin([year])]
print(len(dg_df))
dg_df.head()
```

```python
pop = pop_df[['GeoID', 'County', 'State', f'Pop_{year}']]
print(len(pop))
pop.head()
```

```python
merged_df = pd.merge(dg_df, pop, on='GeoID')
print(len(merged_df))
merged_df.head()
merged_df.columns
```

```python
merged_df=merged_df[['GeoID','County', 'State', 'Year', 
                     'D_Etab', 'D_Emp', 'D_Pay', 'G_Etab', 'G_Emp', 'G_Pay', f'Pop_{year}']]
merged_df.head()
```

```python
merged_df['D_Cap'] = (merged_df[f'Pop_{year}']/merged_df['D_Etab']).astype(int)
merged_df['G_Cap'] = (merged_df[f'Pop_{year}']/merged_df['G_Etab']).astype(int)
merged_df.head()
```

```python
# 2015 prediction coefficients
merged_df[f'FD_num_{year}'] = merged_df.D_Etab*787\
                            + merged_df.D_Emp*165\
                            + merged_df.D_Pay*(-9)\
                            + merged_df.G_Etab*(-266)\
                            + merged_df.G_Emp*27\
                            + merged_df.G_Pay*(-1)\
                            + merged_df.D_Cap*2\
                            + merged_df.G_Cap*1\
                            -15565

merged_df.head()
```

```python
merged_df[f'FD_pct_{year}'] = ((merged_df[f'FD_num_{year}']/merged_df[f'Pop_{year}'])*100).round(2)
merged_df.head()
```
