```python
import pandas
import pandas as pd
import os
```

```python
# read data 2012
year = 2012
df = pd.read_csv(f'{year}_Grocery.csv', encoding='latin-1')
df = df.drop(df.index[0]) # CSV contains two headers. Drop second header.
df_12 = df

year = 2013
df = pd.read_csv(f'{year}_Grocery.csv', encoding='latin-1')
df = df.drop(df.index[0]) # CSV contains two headers. Drop second header.
df_13 = df

year = 2014
df = pd.read_csv(f'{year}_Grocery.csv', encoding='latin-1')
df = df.drop(df.index[0]) # CSV contains two headers. Drop second header.
df_14 = df

year = 2015
df = pd.read_csv(f'{year}_Grocery.csv', encoding='latin-1')
df = df.drop(df.index[0]) # CSV contains two headers. Drop second header.
df_15 = df

year = 2016
df = pd.read_csv(f'{year}_Grocery.csv', encoding='latin-1')
df = df.drop(df.index[0]) # CSV contains two headers. Drop second header.
df_16 = df
```

```python
# Rename columns
df_12 = df_12.rename(columns={"GEO.id2": "GeoID", "GEO.display-label": "County_State", "ESTAB": "Estab"})
# Select Columns
df_12 = df_12[["County_State", "Estab"]]

# Rename columns
df_13 = df_13.rename(columns={"GEO.id2": "GeoID", "GEO.display-label": "County_State", "ESTAB": "Estab"})
# Select Columns
df_13 = df_13[["County_State", "Estab"]]

# Rename columns
df_14 = df_14.rename(columns={"GEO.id2": "GeoID", "GEO.display-label": "County_State", "ESTAB": "Estab"})
# Select Columns
df_14 = df_14[["County_State", "Estab"]]

# Rename columns
df_15 = df_15.rename(columns={"GEO.id2": "GeoID", "GEO.display-label": "County_State", "ESTAB": "Estab"})
# Select Columns
df_15 = df_15[["County_State", "Estab"]]

# Rename columns
df_16 = df_16.rename(columns={"GEO.id2": "GeoID", "GEO.display-label": "County_State", "ESTAB": "Estab"})
# Select Columns
df_16 = df_16[["County_State", "Estab"]]
```

```python
df_12_13 = pd.merge(
    df_12, df_13, on="County_State", suffixes=("_12", "_13"))
df_12_13.head()
```

```python
df_14_15 = pd.merge(
    df_14, df_15, on="County_State", suffixes=("_14", "_15"))
df_14_15.head()
```

```python
df_12_13_14_15 = pd.merge(
    df_12_13, df_14_15, on="County_State")
df_12_13_14_15.head()
```

```python
# Mege data
grocery_df = pd.merge(
    df_12_13_14_15, df_16, on="County_State", suffixes=("", "_16"))
grocery_df.head(20)
```

```python
# Split County & State
grocery_df[['County','State']] = grocery_df.County_State.str.split(",",expand=True,)
grocery_df.head()
```

```python
grocery_df = grocery_df[['County','State', 'Estab_12', 'Estab_13', 'Estab_14', 'Estab_15', 'Estab']]
grocery_df.head()
```

```python
grocery_df.to_csv("Grocery_Output/GroceryData.csv", index=False, header=True)
```
